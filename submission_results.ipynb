# Visualize the prediction errors from the test submission

```python
# 🧩 Step 1: Imports
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import display, HTML

# 🗂️ Step 2: Read the CSV file
df = pd.read_csv('submission_results.csv')

# ❌ Step 3: Filter incorrect predictions
incorrect_df = df[df['CORRECT'] == 0]

# 🔁 Step 4: Display each incorrect prediction with image
for idx, row in incorrect_df.iterrows():
    img_path = row['ID']
    img_path = f"dataset/Test/{img_path}"
    predicted_class = row['CLASS']
    true_class = row['TRUE_CLASS']

    try:
        img = Image.open(img_path)
        plt.figure(figsize=(4, 4))
        plt.imshow(img)
        plt.axis('off')
        plt.title(f'Predicted: {predicted_class} | True: {true_class}')
        plt.show()

        # Add horizontal rule to separate images
        display(HTML("<hr style='border: 1px solid #ccc; margin: 20px 0;'>"))
        
    except Exception as e:
        print(f"Could not display image {img_path}: {e}")
```
